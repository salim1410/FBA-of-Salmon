```python
# Store reactions and their fluxes
def get_gene_reaction_fluxes(solution, gene_id):
    gene_related_reactions = []

    # Iterate through all reactions in the model
    for reaction in model.reactions:
        # Check if the gene is associated with this reaction
        if any(str(gene_id) == str(gene.id) for gene in reaction.genes):
            # Get the flux value for this reaction
            flux_value = solution.fluxes.get(reaction.id, 0)

            gene_related_reactions.append({
                'reaction_id': reaction.id,
                'reaction_name': reaction.name,
                'reaction equation':reaction.reaction,
                'flux_value': flux_value,
                'lower_bound': reaction.lower_bound,
                'upper_bound': reaction.upper_bound,
                'gene_reaction_rule': reaction.gene_reaction_rule
            })

    return {
        'gene_id': gene_id,
        'total_related_reactions': len(gene_related_reactions),
        'reactions': gene_related_reactions
    }

# Example usage
# Assuming 'solution' is your already loaded metabolic solution
gene_ids = ["106561570", "100380817", "100196601", "106613098", "106563252", "106568771", "106568771"]  # Replace with your specific gene ID
for gene_id in gene_ids:
    result = get_gene_reaction_fluxes(solution, gene_id)

    # Print detailed results
    print(f"Gene ID: {result['gene_id']}")
    print(f"Total Related Reactions: {result['total_related_reactions']}")
    print("\nDetailed Reaction Fluxes:")
    for reaction in result['reactions']:
        print("\nReaction ID:", reaction['reaction_id'])
        print("Reaction Name:", reaction['reaction_name'])
        print("Reaction equation:", reaction['reaction equation'])
        print("Flux Value:", reaction['flux_value'])
        print("Lower Bound:", reaction['lower_bound'])
        print("Upper Bound:", reaction['upper_bound'])
        print("Gene Reaction Rule:", reaction['gene_reaction_rule'])
        
    print("\n-----------------------------------------------------------------------\n")
```
